```python
plt.figure(figsize=(15, 30))
pca_features = [f'V{i}' for i in range(1, 29)]

for i, col in enumerate(pca_features, 1):
    plt.subplot(14, 2, i)
    sns.kdeplot(df[df['Class'] == 0][col], label='Legit', fill=True, color='blue', alpha=0.4)
    sns.kdeplot(df[df['Class'] == 1][col], label='Fraud', fill=True, color='red', alpha=0.4)
    plt.title(f'Distribution of {col}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np


X = df.drop(columns=["id", "Class"])
y = df["Class"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA()
pca.fit(X_scaled)

explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio per component:\n", explained_variance)


cumulative_variance = np.cumsum(explained_variance)
print("\nCumulative explained variance:\n", cumulative_variance)


n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"\nNumber of components to reach 95% variance: {n_components_95}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report ,confusion_matrix,roc_auc_score

pca =PCA(n_components=22)
X_pca=pca.fit_transform(X_scaled)

print(f"Shape after PCA: {X_pca.shape}")

```

```python
X_train,X_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=42,stratify=y)
```

```python
model=LogisticRegression(max_iter=1000,class_weight='balanced',random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_prob=model.predict_proba(X_test)[:,1]
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

```

```python
y_pred_dt = dt_model.predict(X_test)
y_prob_dt = dt_model.predict_proba(X_test)[:, 1]

# Evaluation
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob_dt))
```
